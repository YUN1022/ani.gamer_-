# 第一集

## 讀檔案

```python
import pandas as pd 

path = './danmu/112522/'
df = pd.read_csv(path + '112522_1.csv')
df.head()
```

## 以分鐘畫圖

```python
min_interval = []
minute = [i for i in range(1,24)]
for i in range(1, 24):
    f = ((i-1) * 600 <= df['time']) & (df['time'] < i * 600)
    min_interval.append(len(df[f]))

print(min_interval)
print(minute)

df_data = pd.DataFrame({
    'minute':minute,
    'min_interval':min_interval
})
```

```python
sum(min_interval)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.set(style='darkgrid',font_scale=1.5)

g = sns.lineplot(x="minute", y="min_interval",data=df_data)

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.title("以分鐘作為區間")
plt.xlabel("分鐘")
plt.ylabel("彈幕數量")

plt.savefig('分間隔.jpg', dpi=100)
plt.show()
```

# 其他集

```python
path = './112522/'

def draw(index):
    df = pd.read_csv(path + '112522_{}.csv'.format(index))
    df.head()
    
    sec_10_interval = []
    second_10= [i for i in range(1,144)]
    for i in range(1, 144):
        f = ((i-1) * 100 <= df['time']) & (df['time'] < i * 100)
        sec_10_interval.append(len(df[f]))
    
    plt.figure(figsize=(12,8))
    plt.plot(second_10, sec_10_interval)
    plt.savefig("Episode{}_10sec.png".format(index))
    plt.show()
    plt.clf()
```

```python
for i in range(1,17):
    draw(i)
```
