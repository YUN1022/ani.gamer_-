# 讀檔

```python
import pandas as pd 
path = './112522/'
pd.set_option('display.max_rows', None)

all_episode = []
for episode in range(1,17):
    df = pd.read_csv(path + '112522_{}.csv'.format(episode))
    
    sec_10_interval = []
    for i in range(1, 144):
        f = ((i-1) * 100 <= df['time']) & (df['time'] < i * 100)
        sec_10_interval.append(len(df[f]))
    all_episode.append(sec_10_interval)
```

```python
def slope(data, x1, x2):
    y1 = data.loc[x1][0]
    y2 = data.loc[x2][0]
    return (y2-y1) / (x2-x1)
```

# function

```python
def get_Highlight(data ,T):
    
    i = 0
    j = 142
    
    if T > 1:
        left_baseline = slope(data, T-1, T)
        for i in range(T-1, 0, -1):
            if abs(left_baseline - slope(data, T, i)) >= 10:
                break
    
    if T < 142:
        right_baseline = slope(data, T, T+1)
        for j in range(T+1, 142):
            if abs(right_baseline - slope(data, j, T)) >= 10:
                break
    
    return i,j
```

```python
get_Highlight(df_count_7, 64)
```

```python
print(get_Highlight(df_count_7, 64))
print(get_Highlight(df_count_7, 81))
print(get_Highlight(df_count_7, 116))
print(get_Highlight(df_count_7, 121))
print(get_Highlight(df_count_7, 132))
```

```python
def two_std_Highlight(all_episode ,i):
    df_count = pd.DataFrame({
    'count':all_episode[i-1]
    })
    f = df_count['count'] > (np.mean(all_episode[i-1]) - 1) + (2 * np.std(all_episode[i-1]) - 1)
    for time in df_count[f].index:
        return get_Highlight(df_count, time)
```

```python
two_std_Highlight(all_episode,1)
```

```python
for i in range(1, 16):
    two_std_Highlight(all_episode,i)
```
