# 讀檔

```python
import pandas as pd 
path = './112522/'
pd.set_option('display.max_rows', None)

all_episode = []
for episode in range(1,17):
    df = pd.read_csv(path + '112522_{}.csv'.format(episode))
    
    sec_10_interval = []
    for i in range(1, 144):
        f = ((i-1) * 100 <= df['time']) & (df['time'] < i * 100)
        sec_10_interval.append(len(df[f]))
    all_episode.append(sec_10_interval)
```

# 第6集 資料分析

```python
df_count_6 = pd.DataFrame({
    'count':all_episode[5]
})
df_count_6[(df_count_6['count'] > 0)]
```

```python
import numpy as np
print('Max', max(all_episode[5]))
print('Min', min(all_episode[5]))
print('Mean', np.mean(all_episode[5]))
print('Std', np.std(all_episode[5]))
```

## 大於平均數的片段

```python
f = df_count_6['count'] > np.mean(all_episode[5]) - 1
df_count_6[f]
```

## 大於兩個標準差片段

```python
f = df_count_6['count'] > (np.mean(all_episode[5]) - 1) + (2 * np.std(all_episode[5]) - 1)
df_count_6[f]
```
